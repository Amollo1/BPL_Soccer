**Below I have created two more columns. That is (GoalsPerMatch and MinsPerMarch)**

```python
#create two more columns
EPL_df['MinsPerMatch'] = (EPL_df['Mins'] / EPL_df['Matches']).astype(int)
EPL_df['GoalsPerMatch'] = (EPL_df['Goals'] / EPL_df['Matches']).astype(float)
```

```python
EPL_df.head()
```

**Below I am going to calculate the to total number of goals that were scored in 2020 - 2021 epl season**

```python
Total_Goals = EPL_df['Goals'].sum()
```

```python
Total_Goals
```

**Below I want to find out the number of goals that were scored from penalties**

```python
Total_Penalties = EPL_df['Penalty_Goals'].sum()
```

```python
Total_Penalties
```

**Below I want to find the total number of the penalties attempted**

```python
Total_PenaltyAttempts = EPL_df['Penalty_Attempted'].sum()
```

```python
Total_PenaltyAttempts
```

**Below I am going to create a plot of penalties scored agaist the penalties missed in 2020 - 2021 EPL season using pie chart.**

```python
plt.figure(figsize = (13,6))
pl_not_scored = EPL_df['Penalty_Attempted'].sum() - Total_Penalties
data = [pl_not_scored, Total_Penalties]
labels = ['Missed Penalties','Scored Penalties']
color = sns.color_palette('Set2')
plt.pie(data,labels = labels, colors = color, autopct = '%2f%%')
plt.show()
```

**Below I want to display the players in age groups**

```python
Under20 = EPL_df[EPL_df['Age'] <= 20]
Age20_25 = EPL_df[(EPL_df['Age'] > 20) & (EPL_df['Age'] <= 25)]
Age25_30 = EPL_df[(EPL_df['Age'] > 25) & (EPL_df['Age'] <= 30)]
Above30 = EPL_df[EPL_df['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(),Age20_25['Name'].count(),Age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ["Under 20 Years","Between 20 and 25 Years", "Between 25 and 30 Years", "Above 30 Years"]
plt.title('Total Players with Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = "%.1f%%" )
plt.show()
```

**Below I am going to create a plot of the everage age of players in each club**

```python
plt.figure(figsize = (12,6))
sns.boxplot(x = 'Club', y = 'Age', data = EPL_df)
plt.xticks(rotation = 90, fontsize = 15)
```

**Below I am verfifying the average age of players**

**Below we are going to find thetotal assist from each club**

```python
Assists_by_clubs = pd.DataFrame(EPL_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = "whitegrid",color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Assists',data = Assists_by_clubs.sort_values(by = "Assists"),palette = 'Set2')
ax.set_xlabel("Club",fontsize = 30)
ax.set_ylabel("Assists",fontsize = 30)
plt.xticks(rotation = 85, fontsize = 20)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Assits',fontsize = 20)
```

**Below I am going to show the players that gave the most assist in 2020 - 2021 season**

**Below I am going to show the total goals from each club**

```python
Goals_by_clubs = pd.DataFrame(EPL_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style = "whitegrid", color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = Goals_by_clubs.sort_values(by = "Goals"),palette = 'Set2')
ax.set_xlabel("Club", fontsize = 30)
ax.set_ylabel("Goals",fontsize = 20)
plt.xticks(rotation = 75, fontsize = 25)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize = 20)
```
